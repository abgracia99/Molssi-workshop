```python
import os
```

```python
import numpy 
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)            
```

```python
print(headers)
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
sample = headers[1]
sample = headers[2]

plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample)
#adding labels be like
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()
#plt.savefig(F'{sample}.png') #F is string 
plt.savefig ('two_samples.png', dpi=300)
```

```python
num_columns = len(headers)

print(headers)
print(F'The number of columns is {num_columns}')

for col in range(1, num_columns):
    print(F'The column number is {col}.')
    plt.plot(data[:,col], label=headers[col])
    
plt.xlabel('Stimulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()
plt.savefig('all_samples.png', dpi=300)
```

```python
num_columns = len(headers)

print(headers)
print(F'The number of columns is {num_columns}')

for col in range(1, num_columns):
    plt.figure()
    print(F'The column number is {col}.')
    plt.plot(data[:,col], label=headers[col])
    plt.legend()
    plt.xlabel('Stimulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.savefig(F'{headers[col]}.png', dpi=200)
    
```
